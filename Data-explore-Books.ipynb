```python
import seaborn as sns 
```

```python
ax = plt.axes()
sns.heatmap(df.isna().transpose(), cbar=False, ax=ax)

plt.xlabel('columns')
plt.ylabel('Missing Values')

plt.show()
```

```python
import numpy as np
```

```python
df['missing_description'] = np.where(df['description'].isna(),1,0)
df['age_of_books'] = 2024-df['published_year']
```

```python
# columns of interests
columns_of_interest = ['num_pages','age_of_books','missing_description','average_rating']

correlation_matrix = df[columns_of_interest].corr(method = 'spearman')

sns.set_theme(style='white')
plt.figure(figsize=(8,6))
```

```python
heatmap = sns.heatmap(correlation_matrix,annot=True,fmt='.2f',cmap='coolwarm',cbar_kws={"label":"Spearman correlation"})

heatmap.set_title('correlation heatmap')
plt.show()
```

```python
df[(df['description'].isna()) |
    (df['num_pages'].isna()) |
     (df['average_rating'].isna()) |
     (df['published_year'].isna())
    ]
```

```python
df.isnull().sum()
```

```python
# deleting the missing vaiues 

missing_values = df[(~df['description'].isna()) &
     ~(df['num_pages'].isna()) &
     ~(df['average_rating'].isna()) &
     ~(df['published_year'].isna())
    ]
```

```python
missing_values
```

```python
missing_values.isnull().sum()
```

```python
missing_values['categories'].value_counts().reset_index().sort_values('count',ascending=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Prepare the data (Top 20 categories)
top_categories = missing_values['categories'].value_counts().head(20).reset_index()

# Set the style
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 8))

# Create the barplot with the updated syntax
sns.barplot(
    data=top_categories, 
    y='categories', 
    x='count', 
    hue='categories',  # Assign y to hue
    palette='viridis',
    legend=False       # Hide the legend since it's redundant here
)

plt.title('Top 20 Book Categories', fontsize=15)
plt.xlabel('Number of Books', fontsize=12)
plt.ylabel('Category', fontsize=12)
plt.show()
```

```python
book_discription_25_words = missing_values[missing_values['words_in_description'] >= 25]
```

```python
book_discription_25_words
```

```python
book_discription_25_words['title_and_subtitle'] = np.where(
    book_discription_25_words['subtitle'].isna(),
    book_discription_25_words['title'],
    book_discription_25_words['title'].astype(str) + " : " + book_discription_25_words['subtitle'].astype(str)
)
```

```python
book_discription_25_words
```

```python
book_discription_25_words["tagged_description"] = book_discription_25_words[['isbn13','description']].astype(str).agg(" ".join, axis=1)
```
