```python
def generate_prediction(sequence,categories):
    prediction = classifier(sequence,categories)
    max_index = np.argmax(prediction['scores'])
    max_label = prediction['labels'][max_index]
    return max_label
```

```python
from tqdm import tqdm

actual_cats = []
predicted_cats = []

for i in tqdm(range(0,300)):
    sequence = books.loc[books['simple_categories'] == 'Fiction','description'].reset_index(drop=True)[i]
    predicted_cats += [generate_prediction(sequence,Fiction_categories)]
    actual_cats += ['Fiction']
```

```python
for i in tqdm(range(0, 300)):
    sequence = books.loc[books["simple_categories"] == "Nonfiction", "description"].reset_index(drop=True)[i]
    predicted_cats += [generate_prediction(sequence, Fiction_categories)]
    actual_cats += ["Nonfiction"]
```

```python
def generate_prediction_v2(sequence, categories):
    # The hypothesis_template gives the model better context for books
    prediction = classifier(sequence, categories, hypothesis_template="This book is {}")
    max_index = np.argmax(prediction['scores'])
    max_label = prediction['labels'][max_index]
    return max_label
```

```python
# Testing on one of the rows that failed earlier
test_seq = books.loc[books['simple_categories'] == 'Fiction', 'description'].iloc[1]
print(f"Old Label: {generate_prediction(test_seq, Fiction_categories)}")
print(f"New Label: {generate_prediction_v2(test_seq, Fiction_categories)}")
```

```python
# 1. Use the descriptive labels
improved_labels = ["Fictional novel or story", "Non-fiction factual book"]

actual_cats = []
predicted_cats = []

# 2. Run the loop with the v2 function and new labels
for i in tqdm(range(0, 300)):
    # Fiction group
    seq_f = books.loc[books['simple_categories'] == 'Fiction', 'description'].reset_index(drop=True)[i]
    pred_f = generate_prediction_v2(seq_f, improved_labels)
    # Map the long label back to 'fiction' for the report
    predicted_cats.append("fiction" if "Fictional" in pred_f else "nonfiction")
    actual_cats.append("fiction")

for i in tqdm(range(0, 300)):
    # Nonfiction group
    seq_nf = books.loc[books['simple_categories'] == 'Nonfiction', 'description'].reset_index(drop=True)[i]
    pred_nf = generate_prediction_v2(seq_nf, improved_labels)
    # Map back to 'nonfiction'
    predicted_cats.append("fiction" if "Fictional" in pred_nf else "nonfiction")
    actual_cats.append("nonfiction")
```

```python
from sklearn.metrics import classification_report

print("--- Improved Model Performance ---")
print(classification_report(actual_cats, predicted_cats))
```
