```python
import numpy as np
```

```python
products["missing_rating"] = products["rating_score"].isna().astype(int)

products["discount_amount"] = products["price_original"] - products["price_current"]

products["discount_pct"] = np.where(
    products["price_original"] > 0,
    (products["price_original"] - products["price_current"]) / products["price_original"] * 100,
    0
)
```

```python
# column of interest 

columns_of_interest = [

    "price_current",
    "price_original",
    "discount_pct",
    "reviews_count",
    "missing_rating"
]
```

```python
# spearman correlation 
correlation_matrix = products[columns_of_interest].corr(method = 'spearman')
```

```python
# heat map 

sns.set_theme(style = 'white')

plt.figure(figsize=(10,7))

heatmap = sns.heatmap(

    correlation_matrix,
    annot = True,
    fmt = '.2f',
    cmap='coolwarm',
    cbar_kws={'label':"spearman correlation"}
)

heatmap.set_title("Product feature correlation heatmap")
plt.show()
```

```python
products['missing_rating'] = products['rating_score'].isna().astype(int)
```

```python
products_clean = products.dropna(subset=[
    'price_current',
    'category',
    'subcategory'
])
```

```python
products_clean['rating_score'] = products_clean['rating_score'].fillna(
    products_clean['rating_score'].median()
)
```

```python
corr = products_clean.corr(numeric_only=True)
corr["rating_score"].sort_values(ascending=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.show()
```

```python
# Drop columns with zero variance for a cleaner heatmap
corr_clean = corr.dropna(how='all', axis=0).dropna(how='all', axis=1)
sns.heatmap(corr_clean, annot=True, cmap="coolwarm")
```
